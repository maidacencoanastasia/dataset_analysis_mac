```python
import seaborn as sns


def plot(grouped):
    item = grouped["Item"].sum()
    item_list = item.sort_index()
    item_list = item_list[-20:]  # Select top 20 items

    plt.figure(figsize=(9, 10))  # Adjust figure size as needed
    sns.barplot(x=item_list.index, y=item_list.values)  # Pass data directly

    # Optional: Customize labels if needed (similar to your original approach)
    labels = [aj.get_text()[-40:] for aj in plt.gca().get_yticklabels()]  # Using plt.gca()
    plt.gca().set_yticklabels(labels)  # Using plt.gca() to access the current axes

```

```python
sugar = data.groupby(data["Sugars"])
plot(sugar)
```

```python
vitaminC = data.groupby(data["Vitamin C (% Daily Value)"])
plot(vitaminC)
```
